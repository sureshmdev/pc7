```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load your CSV
file_path = "overseas-trade.csv"
df = pd.read_csv(file_path)

# Create dummy sequential & parallel time columns
np.random.seed(42)  
n = len(df)
df["Sequential_time"] = np.random.uniform(0.02, 0.05, n)
df["Parallel_time"] = df["Sequential_time"] + np.random.uniform(-0.01, 0.03, n)

# Calculate total times
total_seq = df["Sequential_time"].sum()
total_par = df["Parallel_time"].sum()

print("Sequential Time:", total_seq)
print("Parallel Time:", total_par)

# Plot overall comparison
plt.figure(figsize=(6, 5))
plt.bar(["Sequential", "Parallel"], [total_seq, total_par], color=["blue", "orange"])
plt.ylabel("Total Time (seconds)")
plt.title("Overall Parallel vs Sequential Time")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
